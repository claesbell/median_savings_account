```python
import pandas as pd
```

```python
data = pd.read_csv('survey-consumer-finances-data.csv')
```

```python
data.info()
```

```python
data['TOTAL_SAVINGS'] = data['SAVING'] + data['MMDA']
```

```python
data_2013 = data[data.YEAR == 2013]
```

```python
data_2013
```

```python
data_2013.info()
```

```python
total_savings_data_2013=data_2013[['CASEID','TOTAL_SAVINGS']]
```

```python
total_savings_data_2013
```

```python
households_with_savings_2013 = total_savings_data_2013[total_savings_data_2013['TOTAL_SAVINGS'] > 0]
```

```python
households_with_savings_2013
```

```python
median_total_savings_2013=households_with_savings_2013.TOTAL_SAVINGS.median()
```

```python
median_total_savings_2013
```
